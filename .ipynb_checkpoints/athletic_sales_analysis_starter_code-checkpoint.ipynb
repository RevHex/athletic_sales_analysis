```python
# Import Libraries and Dependencies
import pandas as pd
```

### 1. Combine and Clean the Data
#### Import CSVs

```python
# Read the CSV files into DataFrames.
as_20_df = pd.read_csv("Resources/athletic_sales_2020.csv")
as_21_df = pd.read_csv("Resources/athletic_sales_2021.csv")
```

```python
# Display the 2020 sales DataFrame
as_20_df
```

```python
# Display the 2021 sales DataFrame
as_21_df
```

#### Check the data types of each DataFrame

```python
# Check the 2020 sales data types.
as_20_df.dtypes
```

```python
# Check the 2021 sales data types.
as_21_df.dtypes
```

#### Combine the sales data by rows.

```python
# Combine the 2020 and 2021 sales DataFrames on the rows.
joined_sales = pd.concat([as_20_df, as_21_df], axis='rows')
joined_sales
```

```python
#  Reset the index.
joined_sales = joined_sales.reset_index(drop=True)
```

```python
# Check if any values are null.
null_values = joined_sales.isnull()
any_null = null_values.any().any()
any_null
```

```python
# Check the data type of each column
joined_sales.dtypes
```

```python
# Convert the "invoice_date" to a datetime datatype
joined_sales['invoice_date'] = pd.to_datetime(joined_sales['invoice_date'])
```

```python
# Confirm that the "invoice_date" data type has been changed.
joined_sales.dtypes
```

### 2. Determine which Region Sold the Most Products

#### Using `groupby`

```python
# Show the number products sold for region, state, and city.
# Rename the sum to "Total_Products_Sold".
rsc_sales_sum = joined_sales.groupby(['region', 'state', 'city']).sum('units_sold').reset_index().rename(columns={'units_sold':'Total_Products_Sold'})

# Show the top 5 results.
top5_sold = rsc_sales_sum.sort_values(by=['Total_Products_Sold'], ascending=False, ignore_index=(True))
top5_sold.head(5)
```

#### Using `pivot_table`

```python
# Show the number products sold for region, state, and city.
# Rename the "units_sold" column to "Total_Products_Sold"
sales_rcs_sum = joined_sales.pivot_table(index = ['region', 'state', 'city'],
                                         values = 'units_sold',
                                         aggfunc = 'sum').reset_index().rename(columns={'units_sold':'Total_Products_Sold'})
# Show the top 5 results.
sold_top5 = sales_rcs_sum.sort_values(by=['Total_Products_Sold'], ascending=False, ignore_index=(True))
top5_sold.head(5)
```

### 3. Determine which Region had the Most Sales

#### Using `groupby`

```python
# Show the total sales for the products sold for each region, state, and city.
# Rename the "total_sales" column to "Total Sales"
rsc_tsales_sum = joined_sales.groupby(['region', 'state', 'city']).sum('total_sales').reset_index().rename(columns={'total_sales':'Total Sales'})

# Show the top 5 results.
top5_tsold = rsc_tsales_sum.sort_values(by=['Total Sales'], ascending=False, ignore_index=(True))
top5_tsold.head(5)
```

#### Using `pivot_table`

```python
# Show the total sales for the products sold for each region, state, and city.
# Optional: Rename the "total_sales" column to "Total Sales"
tsales_rcs_sum = joined_sales.pivot_table(index = ['region', 'state', 'city'],
                                         values = 'total_sales',
                                         aggfunc = 'sum').reset_index().rename(columns={'total_sales':'Total Sales'})
# Show the top 5 results.
tsold_top5 = tsales_rcs_sum.sort_values(by=['Total Sales'], ascending=False, ignore_index=(True))
top5_tsold.head(5)
```

### 4. Determine which Retailer had the Most Sales

#### Using `groupby`

```python
# Show the total sales for the products sold for each retailer, region, state, and city.
# Rename the "total_sales" column to "Total Sales"
rsc_retailer_sum = joined_sales.groupby(['retailer','region', 'state', 'city']).sum('total_sales').reset_index().rename(columns={'total_sales':'Total Sales'})

# Show the top 5 results.
top5_retailer = rsc_retailer_sum.sort_values(by=['Total Sales'], ascending=False, ignore_index=(True))
top5_retailer.head(5)
```

#### Using `pivot_table`

```python
# Show the total sales for the products sold for each retailer, region, state, and city.
# Optional: Rename the "total_sales" column to "Total Sales"
retailer_rsc_sum = joined_sales.pivot_table(index = ['retailer','region', 'state', 'city'],
                                         values = 'total_sales',
                                         aggfunc = 'sum').reset_index().rename(columns={'total_sales':'Total Sales'})
# Show the top 5 results.
retailer_top5 = retailer_rsc_sum.sort_values(by=['Total Sales'], ascending=False, ignore_index=(True))
retailer_top5.head(5)
```

### 5. Determine which Retailer Sold the Most Women's Athletic Footwear

```python
unique_values = joined_sales['product'].unique()
print(unique_values)
```

```python
# Filter the sales data to get the women's athletic footwear sales data.
mwa_retailer_sold = joined_sales[joined_sales["product"].isin(["Women's Athletic Footwear"])]
mwa_retailer_sold
```

```python
mwa_retailer_sold.columns
```

```python
mwa_retailer_sold.dtypes
```

#### Using `groupby`

```python
# Show the total number of women's athletic footwear sold for each retailer, region, state, and city.
# Rename the "units_sold" column to "Womens_Footwear_Units_Sold"
waf_sold = mwa_retailer_sold.groupby(['retailer','region', 'state', 'city']).sum('units_sold').reset_index().rename(columns={'units_sold':'Womens_Footwear_Units_Sold'})

# Show the top 5 results.
top5_waf = waf_sold.sort_values(by=['Womens_Footwear_Units_Sold'], ascending=False, ignore_index=(True))
top5_waf.head(5)
```

#### Using `pivot_table`

```python
# Show the total number of women's athletic footwear sold for each retailer, region, state, and city.
# Rename the "units_sold" column to "Womens_Footwear_Units_Sold"
sold_waf = mwa_retailer_sold.pivot_table(index = ['retailer','region', 'state', 'city'],
                                         values = 'units_sold',
                                         aggfunc = 'sum').reset_index().rename(columns={'units_sold':'Womens_Footwear_Units_Sold'})
# Show the top 5 results.
waf_top5 = sold_waf.sort_values(by=['Womens_Footwear_Units_Sold'], ascending=False, ignore_index=(True))
waf_top5.head(5)
```

### 5. Determine the Day with the Most Women's Athletic Footwear Sales

```python
# Create a pivot table with the 'invoice_date' column is the index, and the "total_sales" as the values.
# Optional: Rename the "total_sales" column to "Total Sales"
mostday_waf = mwa_retailer_sold.pivot_table(index   = 'invoice_date',
                                            values  = 'total_sales',
                                            aggfunc = 'sum').reset_index().rename(columns={'total_sales':'Total Sales'})

# Show the table.
mostday_waf
```

```python
unique_values = mwa_retailer_sold['invoice_date'].unique()
print(unique_values)
```

```python
mostday_waf.info()
```

```python
mostday_waf['invoice_date'] = pd.to_datetime(mostday_waf['invoice_date'])
```

```python
mostday_waf.set_index('invoice_date', inplace=True)
```

```python
# Resample the pivot table into daily bins, and get the total sales for each day.
daily_bins = mostday_waf.resample('D').mean()

# Sort the resampled pivot table in ascending order on "Total Sales".
daily_bins.sort_values(by="Total Sales", ascending=False).head(10)
```

### 6.  Determine the Week with the Most Women's Athletic Footwear Sales

```python
# Resample the pivot table into weekly bins, and get the total sales for each week.
weekly_bins = mostday_waf.resample('W').mean().round(1)

# Sort the resampled pivot table in ascending order on "Total Sales".
weekly_bins.sort_values(by="Total Sales", ascending=False).head(10)
```
